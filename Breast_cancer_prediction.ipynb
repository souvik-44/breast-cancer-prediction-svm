# Breast cancer prediction model using Support Vector Machine(SVM)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.datasets import load_breast_cancer
```

## STEP #1: LOADING THE DATA

```python
cancer = load_breast_cancer()
print(cancer.keys())
```

```python
print(cancer.DESCR)
```

```python
cancer
```

```python
print(cancer['target'])
```

```python
print(cancer['target_names'])
```

```python
print(cancer['feature_names'])
```

```python
cancer['data'].shape
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns= np.append(cancer['feature_names'],['target']))
```

```python
df_cancer.head()
```

## STEP #2: VISUALIZATION OF THE DATA

```python
sns.pairplot(df_cancer,hue ='target',vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
 'mean smoothness'])
```

```python
sns.countplot(df_cancer['target'])
```

```python
sns.scatterplot(x = 'mean area', y = 'mean smoothness',hue = 'target',data = df_cancer)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df_cancer.corr(),annot = True)
```

## STEP 3 : MODEL TRAINING (FINDING A PROBLEM SOLUTION)

```python
X = df_cancer.drop(['target'], axis =1)
y = df_cancer['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train)
```

## STEP 4: EVALUATION OF THE MODEL
#### We need to check the Type I and Type II errors.

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm =confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot= True)
```

### One of the ways to improve the 7 misclassified data is to normalize the data

## STEP #5: IMPROVING THE MODEL

```python
min_train = X_train.min()
```

```python
range_train = (X_train-min_train).max()
```

```python
X_train_scaled = (X_train - min_train)/range_train
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
svc_model.fit(X_train_scaled,y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm, annot= True)
```

## IMPROVING THE MODEL - PART 2

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "C": [0.1, 1, 10, 100],
    "gamma": [1, 0.1, 0.01, 0.001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,grid_predictions)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test,grid_predictions))
```
